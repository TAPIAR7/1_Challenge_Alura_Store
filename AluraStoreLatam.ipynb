### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_total_tienda = tienda['Precio'].sum()
ingreso_total_tienda2 = tienda2['Precio'].sum()
ingreso_total_tienda3 = tienda3['Precio'].sum()
ingreso_total_tienda4 = tienda4['Precio'].sum()
```

```python
print(f"El ingreso total de la tienda 1 es: {ingreso_total_tienda}")
print(f"El ingreso total de la tienda 2 es: {ingreso_total_tienda2}")
print(f"El ingreso total de la tienda 3 es: {ingreso_total_tienda3}")
print(f"El ingreso total de la tienda 4 es: {ingreso_total_tienda4}")
```

# 2. Ventas por categoría

```python
ingreso_total_por_categoria_tienda = tienda.groupby('Categoría del Producto')['Precio'].sum()
ingreso_total_por_categoria_tienda2 = tienda2.groupby('Categoría del Producto')['Precio'].sum()
ingreso_total_por_categoria_tienda3 = tienda3.groupby('Categoría del Producto')['Precio'].sum()
ingreso_total_por_categoria_tienda4 = tienda4.groupby('Categoría del Producto')['Precio'].sum()
```

```python
print(f"El ingreso total por categoría de la tienda 1 es:\n{ingreso_total_por_categoria_tienda}")
print(f"El ingreso total por categoría de la tienda 2 es:\n{ingreso_total_por_categoria_tienda2}")
print(f"El ingreso total por categoría de la tienda 3 es:\n{ingreso_total_por_categoria_tienda3}")
print(f"El ingreso total por categoría de la tienda 4 es:\n{ingreso_total_por_categoria_tienda4}")
```

# 3. Calificación promedio de la tienda


```python
calificacion_promedio_tienda = tienda['Calificación'].mean()
calificacion_promedio_tienda2 = tienda2['Calificación'].mean()
calificacion_promedio_tienda3 = tienda3['Calificación'].mean()
calificacion_promedio_tienda4 = tienda4['Calificación'].mean()
```

```python
print(f"La calificación promedio de la tienda 1 es: {round(calificacion_promedio_tienda, 1)}")
print(f"La calificación promedio de la tienda 2 es: {round(calificacion_promedio_tienda2, 1)}")
print(f"La calificación promedio de la tienda 3 es: {round(calificacion_promedio_tienda3, 1)}")
print(f"La calificación promedio de la tienda 4 es: {round(calificacion_promedio_tienda4, 1)}")
```

# 4. Productos más y menos vendidos

```python
productos_vendidos_tienda = tienda['Producto'].value_counts().reset_index()
productos_vendidos_tienda.columns = ['Producto', 'Cantidad']
producto_mas_vendido_tienda = productos_vendidos_tienda.nlargest(1, 'Cantidad')
producto_menos_vendido_tienda = productos_vendidos_tienda.nsmallest(1, 'Cantidad')

productos_vendidos_tienda2 = tienda2['Producto'].value_counts().reset_index()
productos_vendidos_tienda2.columns = ['Producto', 'Cantidad']
producto_mas_vendido_tienda2 = productos_vendidos_tienda2.nlargest(1, 'Cantidad')
producto_menos_vendido_tienda2 = productos_vendidos_tienda2.nsmallest(1, 'Cantidad')

productos_vendidos_tienda3 = tienda3['Producto'].value_counts().reset_index()
productos_vendidos_tienda3.columns = ['Producto', 'Cantidad']
producto_mas_vendido_tienda3 = productos_vendidos_tienda3.nlargest(1, 'Cantidad')
producto_menos_vendido_tienda3 = productos_vendidos_tienda3.nsmallest(1, 'Cantidad')

productos_vendidos_tienda4 = tienda4['Producto'].value_counts().reset_index()
productos_vendidos_tienda4.columns = ['Producto', 'Cantidad']
producto_mas_vendido_tienda4 = productos_vendidos_tienda4.nlargest(1, 'Cantidad')
producto_menos_vendido_tienda4 = productos_vendidos_tienda4.nsmallest(1, 'Cantidad')
```

```python
print(f"Los productos mas vendidos en la tienda 1 son:\n{producto_mas_vendido_tienda}")
print(f"Los productos mas vendidos en la tienda 2 son:\n{producto_mas_vendido_tienda2}")
print(f"Los productos mas vendidos en la tienda 3 son:\n{producto_mas_vendido_tienda3}")
print(f"Los productos mas vendidos en la tienda 4 son:\n{producto_mas_vendido_tienda4}")

print(f"Los productos menos vendidos en la tienda 1 son:\n{producto_menos_vendido_tienda}")
print(f"Los productos menos vendidos en la tienda 2 son:\n{producto_menos_vendido_tienda2}")
print(f"Los productos menos vendidos en la tienda 3 son:\n{producto_menos_vendido_tienda3}")
print(f"Los productos menos vendidos en la tienda 4 son:\n{producto_menos_vendido_tienda4}")
```

# 5. Envío promedio por tienda

```python
costo_envio_promedio_tienda = tienda['Costo de envío'].mean()
costo_envio_promedio_tienda2 = tienda2['Costo de envío'].mean()
costo_envio_promedio_tienda3 = tienda3['Costo de envío'].mean()
costo_envio_promedio_tienda4 = tienda4['Costo de envío'].mean()
```

```python
import pandas as pd
import folium
from folium.plugins import HeatMap
import matplotlib.pyplot as plt
import numpy as np

tiendas_data = {
    "Tienda 1": tienda,
    "Tienda 2": tienda2,
    "Tienda 3": tienda3,
    "Tienda 4": tienda4,
}

# 1. Creando Heatmaps individuales para cada tienda

for nombre_tienda, df_tienda in tiendas_data.items():
    # Extraer latitud y longitud
    locations = df_tienda[['lat', 'lon']].values.tolist()

    # Crea el Foilum map centrado en la ubicación promedio
    avg_lat = df_tienda['lat'].mean()
    avg_lon = df_tienda['lon'].mean()
    m = folium.Map(location=[avg_lat, avg_lon], zoom_start=12)

    # Crea el HeatMap layer
    HeatMap(locations).add_to(m)

    # Guarda el mapa
    m.save(f"heatmap_{nombre_tienda.replace(' ', '_')}.html")
    print(f"Heatmap guardado para {nombre_tienda} como heatmap_{nombre_tienda.replace(' ', '_')}.html")

print("\nheatmaps separados creados. Abrir en el explorador para visualizar.")

# 2. Crear un Heatmap combinado (opcional - si deseas ver todas las tiendas en un solo mapa)

# Combina todas las ubicaciones de todas las tiendas
all_locations = []
all_lats = []
all_lons = []
for df_tienda in tiendas_data.values():
    all_locations.extend(df_tienda[['lat', 'lon']].values.tolist())
    all_lats.extend(df_tienda['lat'].tolist())
    all_lons.extend(df_tienda['lon'].tolist())

if all_locations:
    # Crea el Foilum map centrado en la ubicación promedio de todas las tiendas
    all_avg_lat = np.mean(all_lats)
    all_avg_lon = np.mean(all_lons)
    m_combined = folium.Map(location=[all_avg_lat, all_avg_lon], zoom_start=11)

    
    # Crea el HeatMap layer
    # Por simplicidad, usamos la misma intensidad para cada ubicación
    # Puedes ajustar esto según tus necesidades (por ejemplo, usando el número de transacciones)
    heatmap_data = [[loc[0], loc[1], 1] for loc in all_locations]  # Intensity of 1 for each location
    HeatMap(heatmap_data).add_to(m_combined)

    # Gaurda el mapa combinado
    m_combined.save("heatmap_combinado_tiendas.html")
    print("Heatmap combinado guardado como heatmap_combinado_tiendas.html")
else:
    print("No hay datos disponibles para crear un heatmap combinado.")

print("\nHeatmap combinado creado (si los datos están disponibles). Abra heatmap_combinado_tiendas.html en su explorador para visualizar.")
```
